Solution for 1 a

```python
import random

num_tosses = 10000
heads_count = 0
tails_count = 0

for _ in range(num_tosses):
    outcome = random.choice(['Heads', 'Tails'])
    if outcome == 'Heads':
        heads_count += 1

    else:
        tails_count += 1

prob_heads = heads_count / num_tosses
prob_tails = tails_count / num_tosses

print(f"Experimental probability of heads: {prob_heads:.4f}")
print(f"Experimental probability of tails: {prob_tails:.4f}")
```

Solution for 1 b

```python
import random

num_rolls = 10000
sum_of_7 = 0

for _ in range(num_rolls):
    dice1 = random.randint(1, 6)
    dice2 = random.randint(1, 6)
    sum_of_boths = dice1 + dice2
    if sum_of_boths == 7:
        sum_of_7 += 1

prob_of_comes_7 = sum_of_7/num_rolls

print(f"Probability of comes 7: {prob_of_comes_7:.4f}")      
```

Solution for 2

```python
import random

def estimate_probability(num_trials=10000):
    success_count = 0

    for _ in range(num_trials):
        found_six = False
        for _ in range(10):  
            roll = random.randint(1, 6)
            if roll == 6:
                found_six = True
                break  
        if found_six:
            success_count += 1

    probability = success_count / num_trials
    print(f"Estimated probability of getting at least one '6' in 10 rolls: {probability:.4f}")

estimate_probability()
```

Solution for Ques. 3

```python
import random

colors = ['red'] * 5 + ['green'] * 7 + ['blue'] * 8
num_trials = 1000


draws = [random.choice(colors) for _ in range(num_trials)]

red_given_prev_blue = 0
prev_blue_count = 0

blue_given_next_red = 0
red_count = 0

for i in range(1, num_trials):
    prev = draws[i - 1]
    curr = draws[i]

    
    if prev == 'blue':
        prev_blue_count += 1
        if curr == 'red':
            red_given_prev_blue += 1

    
    if curr == 'red':
        red_count += 1
        if prev == 'blue':
            blue_given_next_red += 1


P_R_given_Bprev = red_given_prev_blue / prev_blue_count if prev_blue_count else 0
P_Bprev_given_R = blue_given_next_red / red_count if red_count else 0
P_R = draws.count('red') / num_trials
P_Bprev = draws.count('blue') / num_trials


bayes_estimate = (P_Bprev_given_R * P_R) / P_Bprev if P_Bprev else 0


print(f"Estimated P(R | B_prev): {P_R_given_Bprev:.4f}")
print(f"Bayes Theorem Estimate: {bayes_estimate:.4f}")
```

Solution for Ques. 4

```python
import numpy as np


values = [1, 2, 3]
probabilities = [0.25, 0.35, 0.4]


sample = np.random.choice(values, size=1000, p=probabilities)


empirical_mean = np.mean(sample)
empirical_variance = np.var(sample, ddof=0)  
empirical_std_dev = np.std(sample, ddof=0)   

print(f"Empirical Mean: {empirical_mean}")
print(f"Empirical Variance: {empirical_variance}")
print(f"Empirical Standard Deviation: {empirical_std_dev}")
```

Solution for Ques. 5

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon


mean = 5
scale = mean  

samples = np.random.exponential(scale=scale, size=2000)

plt.figure(figsize=(8, 5))

count, bins, ignored = plt.hist(samples, bins=30, density=True, alpha=0.6, color='skyblue', edgecolor='black', label='Histogram')

# Calculate PDF values for the bins
x = np.linspace(0, bins[-1], 1000)
pdf = expon.pdf(x, scale=scale)

plt.plot(x, pdf, 'r-', lw=2, label='Exponential PDF')

plt.title('Histogram and PDF of Exponential Distribution (mean=5)')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.uniform(low=0.0, high=1.0, size=10000)

n = 30
num_samples = 1000
sample_means = [np.mean(np.random.choice(data, size=n, replace=False)) for _ in range(num_samples)]

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title('Uniform Distribution (n=10000)')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.hist(sample_means, bins=30, color='salmon', edgecolor='black')
plt.title('Distribution of Sample Means (n=30, samples=1000)')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
